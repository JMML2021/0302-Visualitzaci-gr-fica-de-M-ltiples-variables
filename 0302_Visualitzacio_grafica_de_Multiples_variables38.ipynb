```python
# Creo Pie Chart para ver distribucion por generos en los ultimo s a√±os

etiquetas = df10anyos.index
explode = [i/24 for i in range(0,len(df10anyos))]

  
fig1, ax1 = plt.subplots(figsize=(10,10))

normdata = mpl.colors.Normalize(min(df10anyos), max(df10anyos))  
colormap = mpl.cm.BuGn         # Esta linea nos permite cambiar el color
colors =colormap(normdata(df10anyos))


ax1.pie(df10anyos, explode=explode, labels=etiquetas, autopct='%.1f%%',
        pctdistance=0.77,
        labeldistance=1.04,
        shadow=True, 
        startangle=10, 
        colors=colors)
ax1.set_title("Porcentaje peliculas √∫ltimos a√±os (1990-2000)",fontsize=18)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

#draw inner circle
centre_circle = plt.Circle((0,0),.48,fc='white')

fig = plt.gcf()

fig.gca().add_artist(centre_circle)

plt.tight_layout()
#plt.legend(title = "Genres:",labels=labels,loc="right",ncol=3)
plt.show()
```
